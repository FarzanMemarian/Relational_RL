```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading text file
with open ("../logs/from_Hinton/3_cntr_dev/cntr_success_stats.txt","r") as f:
    logs = f.readlines()
logs[0].split()

N = len(logs) - len(logs)%10 - 20
mat = np.zeros((N,8))
for line_num in range(N):
    mat[line_num,:] =  list(map(float, logs[line_num+5].split()))[:]
    
down_ratio = 10
mat_down = np.zeros((int(N/down_ratio),2))
for i in range(int(N/down_ratio)):
    mat_down[i,0] = np.mean(mat[down_ratio*i:down_ratio*(i+1),3])
    mat_down[i,1] = np.mean(mat[down_ratio*i:down_ratio*(i+1),4])
```

```python
plt.figure(1)
plt.semilogy(mat_down[:,1])
plt.ylabel("loss (log)")
plt.xlabel("episodes (500)")
plt.title("training cntr, reset, CNN, 9*9 grid")
```

```python
plt.figure(2)
plt.plot(mat_down[:,0])
plt.ylabel("success ratio")
plt.xlabel("episodes (500)")
plt.title("training cntr, reset, CNN, 11*11 grid, 7 obj")
```
